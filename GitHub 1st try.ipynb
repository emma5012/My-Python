```python
df = pd.read_csv('http://stats.moe.gov.tw/files/detail/107/107_student.csv')
```

```python
df.head()
```

```python
colnames = df.loc[1].values
```

```python
df = df[2:]
```

```python
df.head()
```

```python
df.columns = colnames
```

```python
df.head()
```

```python
colnames
```

```python
undergraduate = df[(df['日間∕進修別'] == 'D 日') & (df['等級別'] == 'B 學士')]
```

```python
undergraduate
```

```python
df2=undergraduate.loc[:,'學校名稱':'女生計'].copy()
```

```python
import locale

#atof
```

```python
locale.setlocale(locale.LC_NUMERIC,'zh_TW.UTF-8')
```

```python
df2['男生計'] = df2['男生計'].apply(locale.atof)
```

```python
df2['女生計'] = df2['女生計'].apply(locale.atof)
```

```python
df2['男女比'] = df2.女生計/df2.男生計
```

```python
df2.head()
```

```python
df3 = df2.sort_values(by=['男女比'],ascending = False)
```

```python
df3[:20].plot.bar(x='學校名稱',y=['男女比'])
```

### 函數：$L(x) = x(x+3)(x-7)(x-13) = x^4-17x^3+31x^2+273x$

```python
x = np.linspace(-5,15,300)  
y = x**4-17*x**3+31*x**2+273*x
plt.plot(x,y)
```

```python
def L(x):
    L = x**4-17*x**3+31*x**2+273*x
    return L
def dL(x):
    dL = 4*x**3-51*x**2+62*x+273
    return dL
```

```python
def GradientDescent(x_init,itera,lr):
    x = x_init
    for i in range(itera):      
        x = x - lr * dL(x)
        print("iteration{0}:, x = {1}, y = {2}".format(i,x,L(x)))
        plt.scatter(x,L(x))
```

```python
plt.plot(x,y)
GradientDescent(6,50,0.001)
```

```python
plt.plot(x,y)
GradientDescent(4,50,0.001)
```

```python
plt.plot(x,y)
GradientDescent(3,50,0.001)
```

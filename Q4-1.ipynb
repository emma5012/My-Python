# 資料來源： 歷年中華民國國民出國目的地人數統計
## 資料內容提供91年至106年中華民國國民出國目的地人數統計
## 本次作業分析  民國105和106年，出國目的地為亞洲地區的出國人數成長率

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib as mpl
mpl.rc('font',family='Noto Sans CJK TC')
```

```python
df = pd.read_csv('https://quality.data.gov.tw/dq_download_csv.php?nid=7325&md5_url=c7884ece65970ed288258993eb8ed447') 
```

```python
df
```

```python
# 只分析亞洲地區
df2=df[(df['首站抵達地'] == '亞洲地區')]
```

```python
df3=df2.drop(columns=['首站抵達地','2002', '2003','2004', '2005','2006', '2007','2008', '2009','2010','2011','2012','2013','2014','2015'])
```

```python
df3
```

```python
import locale

# atof
```

```python
locale.setlocale(locale.LC_NUMERIC, '')
```

```python
df3['2016'] = df3['2016'].apply(locale.atof)
df3['2017'] = df3['2017'].apply(locale.atof)
df3['人數成長率'] = df3['2017']/df3['2016']
```

```python
df3.head()
```

```python
df3['國家']=df.細分
df3.plot.bar(x='國家', y=['人數成長率'])
```

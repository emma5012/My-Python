```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

im=np.array(Image.open('racoon.jpg').convert('L'))

rows,cols=im.shape
for i in range(rows):
    for j in range(cols):
        if (im[i,j]<=128):
            im[i,j]=0
        else:
            im[i,j]=1
            
plt.figure("lena")
plt.imshow(img,cmap='gray')
plt.axis('off')
plt.show()
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# 把影像轉成矩陣模式
im=np.array(Image.open('racoon.jpg'))
rows,cols,dims=im.shape

# 加入salt and pepper雜訊 
for i in range(1000):
    x=np.random.randint(0,rows)
    y=np.random.randint(0,cols)
    im[x,y,:]=255 
for i in range(1000):
    x=np.random.randint(0,rows)
    y=np.random.randint(0,cols)
    im[x,y,:]=0
    
plt.imshow(im)
```
